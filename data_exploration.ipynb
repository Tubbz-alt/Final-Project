```python
cacheDirPath = "C:\\Users\\soria\\Repos\\Final-Project\\viz\\static\\res\\"
# print(cacheDirPath)
filess = []
for (root, dirs, files) in os.walk(cacheDirPath):
    for file in files:
#         print(file)
        filess.append(file)
        

        
```

```python
svgs = []
for i, file in enumerate(filess):
    this_svg = {}
    cacheFilePath = cacheDirPath + file
#     print(cacheFilePath)
    glass = file[:-4]
    this_svg['glass_name'] = glass
    print(glass)
    with open(cacheFilePath, encoding='utf-8') as cacheFile:
        html = cacheFile.read()
        svg_soup = bs(html, 'html.parser')
        this_svg['glass'] = svg_soup.find('path', id='glass')['d']
        this_svg['mask'] = svg_soup.find('path', id='mask')['d']
    svgs.append(this_svg)
    
            
```

```python
svgs
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
mongoDb = client.cocktail_db

collection = mongoDb.glass_svgs

collection.insert_many(svgs)
```
