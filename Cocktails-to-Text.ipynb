```python
import csv
import pandas as pd
```

```python
csv = "mr-boston-endClean-lower.csv"
```

```python
df = pd.read_csv(csv)
```

```python
df.head()
```

```python
df = df.fillna("")
```

```python
df.head(12)
```

```python
for i in range(1, 7):
    print(i)
```

```python
drink_names = []
drink_recipes = []
for i in range(len(df)):
    drink = df.iloc[i, :]
    drink_names.append(drink["name"])
    drink_text = drink["name"] + " - "
    drink_text += drink["glass"] +" - "
    for j in range(1, 7):
        if (drink[f"measurement-{j}"] == "" and drink[f"ingredient-{j}"] == ""):
            pass 
        else:
            j_text = drink[f"measurement-{j}"] + " " + drink[f"ingredient-{j}"] + " - "
            drink_text += j_text
    if drink["special ingredients"] == "":
        pass
    else:
        drink_text += "special ingredients: " + drink["special ingredients"] + " - "
    drink_text += drink["instructions"]        
    drink_recipes.append(drink_text)
```

```python
text_df = pd.DataFrame({"name": drink_names, "recipe": drink_recipes})
text_df.head(20)
```

```python
text_df.iloc[1,:]["recipe"]
```

```python
with open("cocktails_text.txt", "w") as text_doc:
    for cocktail_data in text_df["recipe"]:
        text_doc.write(cocktail_data + "\n")
```
